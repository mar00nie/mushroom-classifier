```python
# store edible mushrooms in a list
root = "edible/"
edible_list = []
path = 'http://www.mushroom.world/data/fungi/'
for file in os.listdir(root):
    file = file.split('_')
    if len(file) == 3:
        file = file[2].split('-thumb')
        if len(file) == 2:
            file[1] = '.jpg'
            file = file[0] + file[1]
        else:
            file = file[0]
            file = file.split('.')
            file[1] = '.jpg'
            file = file[0] + file[1]
        edible_list.append(ClImage(url=path+file, concepts=['edible', 'non-poisonous'], 
                                       not_concepts=['inedible', 'poisonous']))
```

```python
# store inedible mushrooms in a list
root = "inedible/"
inedible_list = []
path = 'http://www.mushroom.world/data/fungi/'
for file in os.listdir(root):
    file = file.split('_')
    if len(file) == 3:
        file = file[2].split('-thumb')
        if len(file) == 2:
            file[1] = '.jpg'
            file = file[0] + file[1]
        else:
            file = file[0]
            file = file.split('.')
            file[1] = '.jpg'
            file = file[0] + file[1]
        inedible_list.append(ClImage(url=path+file, concepts=['inedible', 'non-poisonous'], 
                                         not_concepts=['edible', 'poisonous']))
```

```python
# store poisonous mushrooms in a list
root = "poisonous/"
poison_list = []
path = 'http://www.mushroom.world/data/fungi/'
for file in os.listdir(root):
    file = file.split('_')
    if len(file) == 3:
        file = file[2].split('-thumb')
        if len(file) == 2:
            file[1] = '.jpg'
            file = file[0] + file[1]
        else:
            file = file[0]
            file = file.split('.')
            file[1] = '.jpg'
            file = file[0] + file[1]
        poison_list.append(ClImage(url=path+file, concepts=['inedible', 'poisonous'], 
                                         not_concepts=['edible', 'non-poisonous']))
```
